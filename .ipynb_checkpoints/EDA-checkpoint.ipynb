```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('facemask.csv')
# data guide: spreadsheet_name, spreadsheet_row and trial_number are 
# most likely things you will not be touching during your analysis
df.drop(['spreadsheet_name', 'spreadsheet_row', 'trial_number','participant_device_type', 'participant_browser',
         ]
        , axis=1, inplace=True)
df
```

```python
data = df.groupby(['participant_private_id','condition'])['correct'].value_counts().rename('true_answer').reset_index()
i = data[data.correct == False].index
data = data.drop(i)
data.drop(columns = ['correct'], inplace = True)
data['false_answer'] = 80 - data['true_answer']
data['loudness_accuracy'] = df.groupby(['participant_private_id','condition'])['loudness_accuracy'].mean().reset_index()['loudness_accuracy']
#data['mask_before'] = df.groupby(['participant_private_id','condition'])['mask_before'].max().reset_index()['mask_before']
#data['mask_after'] = df.groupby(['participant_private_id','condition'])['mask_after'].max().reset_index()['mask_after']
#data['mask_people'] = df.groupby(['participant_private_id','condition'])['mask_people'].max().reset_index()['mask_people']
#data['mask_perception'] = df.groupby(['participant_private_id','condition'])['mask_perception'].max().reset_index()['mask_perception']
#data['mask_spread'] = df.groupby(['participant_private_id','condition'])['mask_spread'].max().reset_index()['mask_spread']
#data['mask_freedom'] = df.groupby(['participant_private_id','condition'])['mask_freedom'].max().reset_index()['mask_freedom']
#data['mask_vulnerable'] = df.groupby(['participant_private_id','condition'])['mask_vulnerable'].max().reset_index()['mask_vulnerable']
#data['vaccine'] = df.groupby(['participant_private_id','condition'])['vaccine'].max().reset_index()['vaccine']
#data['intensity'] = df.groupby(['participant_private_id','condition'])['intensity'].sum().reset_index()['intensity']
data['true_rate'] = data['true_answer']/80
data['false_rate'] = data['false_answer']/80
data
```
